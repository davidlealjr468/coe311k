```python
import numpy as np
import time

#The Jacobi Method
def jacobi(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        
        error = np.linalg.norm(x_new - x, ord=np.inf)
        
        if error < tol:
            return x_new, k+1, True, error
        
        x = x_new.copy()
    
    return x, max_iterations, False, error


#The Gauss-Seidel Method
def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_old = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x[j] for j in range(i))
            s2 = sum(A[i][j] * x_old[j] for j in range(i+1, n))
            x[i] = (b[i] - s1 - s2) / A[i][i]
        
        error = np.linalg.norm(x - x_old, ord=np.inf)
        
        if error < tol:
            return x, k+1, True, error
    
    return x, max_iterations, False, error



#Matrices

# Challenge 6a (diagonally dominant)
A6a = np.array([
    [10, -1, 2, 0, 0],
    [-1, 11, -1, 3, 0],
    [2, -1, 10, -1, 0],
    [0, 3, -1, 8, -2],
    [0, 0, 0, -2, 9]
], dtype=float)

b6a = np.array([14, 30, 26, 25, 37], dtype=float)

# Challenge 6b (not diagonally dominant)
A6b = np.array([
    [1, 2, 3, 0, 0],
    [2, 1, 2, 3, 0],
    [3, 2, 1, 2, 3],
    [0, 3, 2, 1, 2],
    [0, 0, 3, 2, 1]
], dtype=float)

b6b = np.array([14, 22, 33, 26, 22], dtype=float)

x0 = np.zeros(5)
tol = 1e-6
max_iter = 100


#Run and Compare
def run_test(A, b, label):
    print(f"\n----- {label} -----")
    
    start = time.time()
    sol_j, it_j, conv_j, err_j = jacobi(A, b, x0, tol, max_iter)
    time_j = time.time() - start
    
    start = time.time()
    sol_gs, it_gs, conv_gs, err_gs = gauss_seidel(A, b, x0, tol, max_iter)
    time_gs = time.time() - start
    
    print("Jacobi:")
    print("  Converged:", conv_j)
    print("  Iterations:", it_j)
    print("  Final Error:", err_j)
    print("  Time:", time_j)
    
    print("\nGauss-Seidel:")
    print("  Converged:", conv_gs)
    print("  Iterations:", it_gs)
    print("  Final Error:", err_gs)
    print("  Time:", time_gs)


run_test(A6a, b6a, "Challenge 6a")
run_test(A6b, b6b, "Challenge 6b")
```
