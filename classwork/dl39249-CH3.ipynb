```python
import numpy as np
import matplotlib.pyplot as plt
import random as r

pi_value = np.pi
true_error = []
relative_error = []

# Monte Carlo Approximation
def approximate_pi(samples):
    count = 0
    for i in range(samples):
        x = r.uniform(-1, 1)
        y = r.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            count += 1
    return 4 * count / samples   

#Runs and trial
num_trials = 100
samples = 1000

for _ in range(num_trials):
    pi_est = approximate_pi(samples)
    
    t_error = abs(pi_value - pi_est)
    r_error = abs((pi_value - pi_est) / pi_value)
    
    true_error.append(t_error)
    relative_error.append(r_error)
    
#Plot
plt.figure()
plt.plot(true_error, label="True Error")
plt.plot(relative_error, label="Relative Error")
plt.xlabel("Trial")
plt.ylabel("Error")
plt.legend()
plt.show()
```
