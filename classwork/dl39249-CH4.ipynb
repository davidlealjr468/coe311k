```python
#Approximation for pi
import numpy as np

term = int(input("Please give me the term for the Leibniz series:"))
pi_value = np.pi

seq = 0
for i in range(term):
    seq += ((-1)**i)/(2*i+1)
    
    if (i+1) % 100 == 0:
        pi_approx = 4*seq
        relative_error = abs(pi_approx - pi_value) / pi_value
        print(f"After {i+1} terms:")
        print(f"  Approximated π =", pi_approx)
        print(f"  Relative Error =", relative_error)
        print("-" * 40)
```

As each term is added to the summation, the error gets smaller
