```python
import numpy as np
import matplotlib.pyplot as plt

#Part 1

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print("PART 1")
print(f"Slope: {m}")
print(f"Intercept: {c}")

plt.figure()
plt.scatter(X, Y)
plt.plot(X, m*X + c)
plt.title("Part 1: Original Data + Least Squares Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()


#Part 2

np.random.seed(0)
X_rand = np.random.permutation(X)
Y_rand = np.random.permutation(Y)

A_rand = np.vstack([X_rand, np.ones(len(X_rand))]).T
m_rand, c_rand = np.linalg.lstsq(A_rand, Y_rand, rcond=None)[0] #Setting the 1st output 

print("\nPART 2")
print(f"Slope: {m_rand}")
print(f"Intercept: {c_rand}")

plt.figure()
plt.scatter(X_rand, Y_rand)
plt.plot(X_rand, m_rand*X_rand + c_rand)
plt.title("Part 2: Randomized Data + Least Squares Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()


#Part 3

# Original equation
m_true = 2.5
c_true = 1.0

x = np.arange(0, 1.01, 0.01)
noise = np.random.normal(0, 0.2, len(x))
y = m_true*x + c_true + noise

A3 = np.vstack([x, np.ones(len(x))]).T
m_fit, c_fit = np.linalg.lstsq(A3, y, rcond=None)[0]

print("\nPART 3")
print("True slope:", m_true)
print("True intercept:", c_true)
print("Fitted slope:", m_fit)
print("Fitted intercept:", c_fit)

plt.figure()
plt.scatter(x, y)
plt.plot(x, m_true*x + c_true)
plt.plot(x, m_fit*x + c_fit)
plt.title("Part 3: True Line vs Least Squares Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
