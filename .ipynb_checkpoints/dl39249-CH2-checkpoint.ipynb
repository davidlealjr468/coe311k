I utilized a variable and implemented it into a for loop
in order to satisfy the user's need on how many numbers they would like to
convert to binary. Before the loop, I made an empty numpy array that way I can add my binary conversions one by one 
into the matrix itself.

```python
import numpy as np

user = int(input("How many numbers would you like to convert to binary?: "))
matrix = np.empty(user)

for i in range(user):
    nums = int(input(f"Give me the number {i+1}: "))
    bit = nums % 2
    if bit == 0: bit = 0
    if bit > 0 and bit < 1 : bit = 1
    matrix[i] = bit 

print(matrix)
```
