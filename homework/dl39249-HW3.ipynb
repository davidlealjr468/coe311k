```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Taylor Series Function (variable number of terms)
def taylor_sin(x, n_terms):
    approximation = 0
    for i in range(n_terms):
        approximation += ((-1)**i) * (x**(2*i+1)) / math.factorial(2*i+1)
    return approximation


# Generate x values
x_values = np.arange(0, 10.1, 0.1)

# Exact values
exact_values = np.sin(x_values)

# Taylor approximations (1â€“4 terms)
approx_1 = np.array([taylor_sin(x, 1) for x in x_values])
approx_2 = np.array([taylor_sin(x, 2) for x in x_values])
approx_3 = np.array([taylor_sin(x, 3) for x in x_values])
approx_4 = np.array([taylor_sin(x, 4) for x in x_values])

# Absolute errors
error_1 = np.abs(exact_values - approx_1)
error_2 = np.abs(exact_values - approx_2)
error_3 = np.abs(exact_values - approx_3)
error_4 = np.abs(exact_values - approx_4)

# Plot truncation error
plt.figure()
plt.plot(x_values, error_1)
plt.plot(x_values, error_2)
plt.plot(x_values, error_3)
plt.plot(x_values, error_4)

plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Truncation Error of Taylor Series Approximation for sin(x)")
plt.legend(["1 term", "2 terms", "3 terms", "4 terms"])
plt.show()

```

Discussion:
The truncation error decreases as more Taylor series terms are included.
However, the decrease is not uniform across all x-values.
The approximation is very accurate near x = 0 because the Taylor series
is centered at zero. As x increases, the truncation error grows,
especially when using only a few terms.
Larger x-values require more terms for accurate approximation.
