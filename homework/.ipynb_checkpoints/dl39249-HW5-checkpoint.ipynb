```python
import numpy as np
import matplotlib.pyplot as plt

#Custom

def my_least_squares(X, Y):
    A = np.vstack([X, np.ones(len(X))]).T
    ATA = A.T @ A
    ATY = A.T @ Y
    beta = np.linalg.inv(ATA) @ ATY
    m, b = beta[0], beta[1]
    return m, b


#Part 1

np.random.seed(1)

X_rand = np.random.rand(20)
Y_rand = np.random.rand(20)

# Sort X (important)
idx = np.argsort(X_rand)
X_rand = X_rand[idx]
Y_rand = Y_rand[idx]

# Custom LS
m_custom, b_custom = my_least_squares(X_rand, Y_rand)

# NumPy LS
A_rand = np.vstack([X_rand, np.ones(len(X_rand))]).T
m_np, b_np = np.linalg.lstsq(A_rand, Y_rand, rcond=None)[0]

print("RANDOM DATA TEST")
print("Custom LS:  m =", m_custom, " b =", b_custom)
print("NumPy  LS:  m =", m_np, " b =", b_np)


#Part 2

# Random slope and intercept
m_true = np.random.uniform(-5, 5)
b_true = np.random.uniform(-2, 2)

X = np.arange(0, 1.01, 0.01)
noise = np.random.uniform(0, 1, len(X))
sign = np.random.choice([-1, 1], len(X))

Y = m_true*X + b_true + sign*noise

# Custom LS
m_fit, b_fit = my_least_squares(X, Y)

# NumPy LS
A = np.vstack([X, np.ones(len(X))]).T
m_np2, b_np2 = np.linalg.lstsq(A, Y, rcond=None)[0]

print("\nRANDOM LINE + NOISE TEST")
print("True m:", m_true, " True b:", b_true)
print("Custom LS m:", m_fit, " b:", b_fit)
print("NumPy  LS m:", m_np2, " b:", b_np2)


#Residual Analysis

Y_pred = m_fit*X + b_fit
residuals = Y - Y_pred

max_res_idx = np.argmax(np.abs(residuals))
min_res_idx = np.argmin(np.abs(residuals))

print("\nResidual Analysis")
print("Largest Residual:", residuals[max_res_idx], " at X =", X[max_res_idx])
print("Smallest Residual:", residuals[min_res_idx], " at X =", X[min_res_idx])


#Plot Results
plt.figure()
plt.scatter(X, Y)
plt.plot(X, m_true*X + b_true)
plt.plot(X, m_fit*X + b_fit)
plt.title("Custom Least Squares vs True Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
