```python
import numpy as np
import matplotlib.pyplot as plt
import time


A = np.array([[3.0, -0.1, -0.2],
              [0.1, 7.0, -0.3],
              [0.3, -0.2, 10.0]], dtype=float)

b = np.array([7.85, -19.3, 71.4], dtype=float)

x0 = np.zeros(3)
maxit = 500
tolerances = [1e-3, 1e-6, 1e-9]



def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    errors = []

    for k in range(max_iterations):
        x_new = np.zeros_like(x)

        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]

        error = np.linalg.norm(x_new - x, ord=np.inf)
        errors.append(error)

        if error < tol:
            return x_new, k+1, errors

        x = x_new

    return x, max_iterations, errors




def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    errors = []

    for k in range(max_iterations):
        x_old = x.copy()

        for i in range(n):
            s1 = sum(A[i][j] * x[j] for j in range(i))
            s2 = sum(A[i][j] * x_old[j] for j in range(i+1, n))
            x[i] = (b[i] - s1 - s2) / A[i][i]

        error = np.linalg.norm(x - x_old, ord=np.inf)
        errors.append(error)

        if error < tol:
            return x, k+1, errors

    return x, max_iterations, errors




iterationsJacobi = []
iterationsGauss = []
errorsJacobi = []
errorsGauss = []

print("Condition Number:", np.linalg.cond(A))

for tol in tolerances:
    print(f"\nTolerance = {tol}")

    start = time.time()
    solJ, itJ, errJ = jacobi(A, b, x0, tol, maxit)
    timeJ = time.time() - start

    start = time.time()
    solG, itG, errG = gauss_seidel(A, b, x0, tol, maxit)
    timeG = time.time() - start

    print(f"Jacobi:        {itJ} iterations, time = {timeJ:.6f}s")
    print(f"Gauss-Seidel:  {itG} iterations, time = {timeG:.6f}s")

    iterationsJacobi.append(itJ)
    iterationsGauss.append(itG)
    errorsJacobi.append(errJ)
    errorsGauss.append(errG)


#Plotting
plt.figure(figsize=(12, 6))

for i in range(3):
    # Jacobi (top row)
    plt.subplot(2, 3, i+1)
    plt.semilogy(errorsJacobi[i])
    plt.title(f"Jacobi (tol={tolerances[i]})")
    plt.xlabel("Iteration")
    plt.ylabel("Error")
    plt.grid(True)

    # Gauss-Seidel (bottom row)
    plt.subplot(2, 3, i+4)
    plt.semilogy(errorsGauss[i])
    plt.title(f"Gauss-Seidel (tol={tolerances[i]})")
    plt.xlabel("Iteration")
    plt.ylabel("Error")
    plt.grid(True)

plt.tight_layout()
plt.show()
```
