```python
import numpy as np
import math

term = 4   # number of Taylor terms

x_values = np.arange(0, 10.1, 0.1)

for x in x_values:
    
    # Exact value
    exact = np.sin(x)
    print("Sin value", x, ":", exact)
    
    # Taylor approximation
    seq = 0
    for i in range(term):
        seq += ((-1)**i) * (x**(2*i+1)) / math.factorial(2*i+1)
    
    print("Taylor approximation:", seq)
    print("Absolute error:", abs(exact - seq))
    print("-------------------------------")
    


```
