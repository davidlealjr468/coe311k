```python
import numpy as np
import matplotlib.pyplot as plt
import random as r


x = np.linspace(10, 1000, 20, dtype = int)

pi_value = np.pi
errors = []

# Monte Carlo Approximation
def approximate_pi(samples):
    count = 0
    for i in range(samples):
        x = r.uniform(-1, 1)
        y = r.uniform(-1, 1)
        if x**2 + y**2 < 1:
            count += 1
    return 4 * count / samples   

#Error for each approximation
for samples in x:
    pi_est = approximate_pi(samples)
    error = abs(pi_value - pi_est)
    errors.append(error)

#Plot
plt.figure()
plt.title("Monte Carlo Error")
plt.xlabel("Number of Samples")
plt.ylabel("Error")
plt.plot(x, errors)
plt.show()
```
