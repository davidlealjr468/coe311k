```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x**2

plt.subplot(221)
plt.plot(x, y1, 'r')
plt.title("sin")

plt.subplot(222)
plt.plot(x, y2, 'r')
plt.title("cosine")

plt.subplot(223)
plt.plot(x, y3, 'r')
plt.title("tangent")

plt.subplot(224)
plt.plot(x, y4, 'r')
plt.title("Quadratic")

plt.tight_layout()   # adjusts spacing between all 4 plots
plt.show()
```
